# 03 — Bayesian Optimization Demo (placeholder)
Sketch a simple bandit loop. Replace with BoTorch for real GP-based acquisition.

```python
import numpy as np, pandas as pd
from triality_qutrit.optimize import SimpleBandit
df = pd.read_csv('../data/synthetic/triad_sweep.csv')
xs = sorted(df['P_ge_dBm'].unique())
ys = sorted(df['P_ef_dBm'].unique())
bandit = SimpleBandit(xs, ys)
def oracle(xy):
    sub = df[(df.P_ge_dBm==xy[0]) & (df.P_ef_dBm==xy[1])]
    return float(sub['T2star_ns'].mean())
for _ in range(3):
    for xy in bandit.suggest(10):
        bandit.observe(xy, oracle(xy))
len(bandit.measured)
```
